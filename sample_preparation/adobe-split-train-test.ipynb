```python
import os
from random import shuffle
```

```python
dataFolder = '/home/winston/data/Adobe-240-fps'
videoList = os.listdir(dataFolder)
print("Video number:", len(videoList))
frameNum = 0
for vid in videoList:
    frameNum += len(os.listdir(os.path.join(dataFolder, vid)))
print("Total frames", frameNum)
shuffle(videoList)
```

```python
trainFrameNum = 0
if os.path.exists('adobe-split.txt'):
    os.remove('adobe-split.txt')
for vid in videoList:
    currentFrameNum = len(os.listdir(os.path.join(dataFolder, vid)))
    if trainFrameNum < 0.79*frameNum:
        with open('adobe-split.txt', 'a') as f:
            f.write('train Adobe-240-fps/%s %04d\n' % (vid, currentFrameNum))
        print('train %s %04d' % (vid, currentFrameNum))
        trainFrameNum += currentFrameNum
    else:
        with open('adobe-split.txt', 'a') as f:
            f.write('test Adobe-240-fps/%s %04d\n' % (vid, currentFrameNum))
        print('test %s %04d' % (vid, currentFrameNum))
print("training frame number:", trainFrameNum)
```
