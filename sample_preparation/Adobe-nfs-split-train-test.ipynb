```python
import os
from random import shuffle
from skimage.io import imread
import numpy as np
```

```python
dataFolderNfS = 'F:\winston\data\\need-for-speed'
videoListNfS = os.listdir(dataFolderNfS)
print("Need-for-speed:")
print("--Video number:", len(videoListNfS))
frameNumNfS = 0
for vid in videoListNfS:
    frameNumNfS += len(os.listdir(os.path.join(dataFolderNfS, vid, '240', vid)))
print("--Total frames", frameNumNfS)
clipsNumNfS = frameNumNfS // 50
shuffle(videoListNfS)
```

```python
dataFolderAdobe = 'F:\winston\data\\Adobe-240-fps'
videoListAdobe = os.listdir(dataFolderAdobe)
print("Adobe-240-fps:")
print("--Video number:", len(videoListAdobe))
frameNumAdobe = 0
for vid in videoListAdobe:
    frameNumAdobe += len(os.listdir(os.path.join(dataFolderAdobe, vid)))
print("--Total frames", frameNumAdobe)
clipsNumAdobe = frameNumAdobe // 50
shuffle(videoListAdobe)
```

```python
trainClipNum = 0
testClipNum = 0
# process Adobe data
if os.path.exists('adobe-nfs-split.txt'):
    os.remove('adobe-nfs-split.txt')
for vid in videoListAdobe:
    currentFrameNum = len(os.listdir(os.path.join(dataFolderAdobe, vid)))
    if currentFrameNum > 50:
        sampImg = imread(os.path.join(dataFolderAdobe, vid, '0000.jpg'))
        sz_0 = np.size(sampImg, 0)
        sz_1 = np.size(sampImg, 1)
        if trainClipNum < 0.8*clipsNumAdobe:
            with open('adobe-nfs-split.txt', 'a') as f:
                for idx in range(0, currentFrameNum-50, 50):
                    f.write('train %s %04d %04d %05d\n' % (os.path.join(dataFolderAdobe, vid), sz_0, sz_1, idx))
                    print('train %s %04d %04d %05d\n' % (os.path.join(dataFolderAdobe, vid), sz_0, sz_1, idx))
                    trainClipNum += 1
        else:
            with open('adobe-nfs-split.txt', 'a') as f:
                for idx in range(0, currentFrameNum-50, 50):
                    f.write('test %s %04d %04d %05d\n' % (os.path.join(dataFolderAdobe, vid), sz_0, sz_1, idx))
                    print('test %s %04d %04d %05d\n' % (os.path.join(dataFolderAdobe, vid), sz_0, sz_1, idx))
                    testClipNum += 1
    else:
        print("Error: video length not enough! (at least 50 frames) in", vid)

trainClipNumAdobe = trainClipNum
testClipNumAdobe = testClipNum
        
# process nfs data
trainClipNum = 0
testClipNum = 0
for vid in videoListNfS:
    currentFrameNum = len(os.listdir(os.path.join(dataFolderNfS, vid, '240', vid)))
    if currentFrameNum > 100:
        sampImg = imread(os.path.join(dataFolderNfS, vid, '240', vid, '00001.jpg'))
        sz_0 = np.size(sampImg, 0)
        sz_1 = np.size(sampImg, 1)
        if trainClipNum < 0.8*clipsNumNfS:
            with open('adobe-nfs-split.txt', 'a') as f:
                for idx in range(51, currentFrameNum-50, 50):
                    f.write('train %s %04d %04d %05d\n' % (os.path.join(dataFolderNfS, vid, '240', vid), sz_0, sz_1, idx))
                    print('train %s %04d %04d %05d\n' % (os.path.join(dataFolderNfS, vid, '240', vid), sz_0, sz_1, idx))
                    trainClipNum += 1
        
        else:
            with open('adobe-nfs-split.txt', 'a') as f:
                for idx in range(51, currentFrameNum-50, 50):
                    f.write('test %s %04d %04d %05d\n' % (os.path.join(dataFolderNfS, vid, '240', vid), sz_0, sz_1, idx))
                    print('test %s %04d %04d %05d\n' % (os.path.join(dataFolderNfS, vid, '240', vid), sz_0, sz_1, idx))
                    testClipNum += 1
    else:
        print("Error: video length not enough! (at least 50 frames) in", vid)
print("NfS training clip number:", trainClipNum)
print("Adobe training clip number:", trainClipNumAdobe)
print("Total training clip number:", trainClipNum + trainClipNumAdobe)

print("NfS testing clip number:", testClipNum)
print("Adobe testing clip number:", testClipNumAdobe)
print("Total testing clip number:", testClipNum + testClipNumAdobe)
```
