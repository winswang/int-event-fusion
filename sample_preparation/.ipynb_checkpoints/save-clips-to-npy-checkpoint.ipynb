```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
import tensorflow as tf
import os
import random
```

```python
def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))
def float_list_feature(value):
    return tf.train.Feature(float_list=tf.train.FloatList(value=value))

def get_example_nums(tf_records_filenames):
    '''
    统计tf_records图像的个数(example)个数
    :param tf_records_filenames: tf_records文件路径
    :return:
    '''
    nums= 0
    for record in tf.python_io.tf_record_iterator(tf_records_filenames):
        nums += 1
    return nums

def show_image(title,image):
    '''
    display images
    :param title: 图像标题
    :param image: 图像的数据
    :return:
    '''
    plt.imshow(image)
    plt.axis('on')    # 关掉坐标轴为 off
    plt.title(title)  # 图像题目
    plt.show()
    
def load_split_file(filename,shuffle=False):
    '''
    read in split file, 'train/test /file/path sz_0 sz_1 num_frames'
    :param filename:
    :param shuffle :是否打乱顺序
    :return:images type->list
    :return:num_frames type->list
    '''
    images = []
    sz0 = []
    sz1 = []
    start_frame = []
    train_test = []
    with open(filename) as f:
        lines_list = f.readlines()
        if shuffle:
            random.shuffle(lines_list)

        for lines in lines_list:
            line = lines.rstrip('\n').split(' ')
            if line[0] == 'train':
                train_test.append(int(1))
            else:
                train_test.append(int(0))
            images.append(line[1])
            sz0.append(int(line[2]))
            sz1.append(int(line[3]))
            start_frame.append(int(line[4]))

    return train_test, images, start_frame, sz0, sz1

def read_image(filename, normalization=True):
    '''
    读取图片数据,默认返回的是float[0,1]
    :param filename:
    :param normalization:是否归一化到[0.,1.0]
    :return: image, img_sz
    ''' 
    image = np.array(imread(filename))
    if normalization:
        image = image / 255.0
    return image

def load_clip(img_dir, start_idx, verbose = 0):
    if 'Adobe' in img_dir:
        vid = np.array([read_image(os.path.join(img_dir, '%04d.jpg' % (start_idx + i))) for i in range(50)])
    else:
        vid = np.array([read_image(os.path.join(img_dir, '%05d.jpg' % (start_idx + i))) for i in range(50)])
    if verbose:
        print(np.shape(vid))
    return vid
```

```python
# create records
def create_records(info_file, output_record_dir, resize_height, resize_width, shuffle = False,log=5):
    '''
    实现将图像原始数据,label,长,宽等信息保存为record文件
    注意:读取的图像数据默认是float,再转为tf的字符串型BytesList保存,解析请需要根据需要转换类型
    :param info_file:输入保存图片信息的txt文件
    :param output_record_dir:保存record文件的路径
    :param shuffle:是否打乱顺序
    :param log:log信息打印间隔
    '''
    # load data-info file
    train_test, images_list, start_frame_list, sz0_list, sz1_list = load_split_file(info_file, shuffle)

    writer = tf.python_io.TFRecordWriter(output_record_dir)

    for i, [image_dir, start_frame] in enumerate(zip(images_list, start_frame_list)):
        clip = load_clip(image_dir, start_frame)
        clip_raw = clip.tostring()

        if i%log==0 or i==len(images_list)-1:
            print('------------processing:%d-th------------' % (i))
            print('current image_path=%s' % (image_dir),'shape:{}'.format(clip.shape))
        example = tf.train.Example(features=tf.train.Features(feature={
            'clip_raw': _bytes_feature(clip_raw),
            'sz0': _int64_feature(sz0_list[i]),
            'sz1': _int64_feature(sz1_list[i]),
            'depth': _int64_feature(50),
            'label': _int64_feature(train_test[i])
        }))
        writer.write(example.SerializeToString())
    writer.close()
    
# read records
def read_records(filename,resize_height, resize_width,type=None):
    '''
    解析record文件:源文件的图像数据是RGB,uint8,[0,255],一般作为训练数据时,需要归一化到[0,1]
    :param filename:
    :param type:选择图像数据的返回类型
         None:默认将uint8-[0,255]转为float32-[0,255]
         normalization:归一化float32-[0,1]
         centralization:归一化float32-[0,1],再减均值中心化
    :return:
    '''
    # 创建文件队列,不限读取的数量
    filename_queue = tf.train.string_input_producer([filename])
    
    # create a reader from file queue
    reader = tf.TFRecordReader()
    
    # reader从文件队列中读入一个序列化的样本
    _, serialized_example = reader.read(filename_queue)

    # get feature from serialized example
    features = tf.parse_single_example(
        serialized_example,
        features={
            'clip_raw': tf.FixedLenFeature([], tf.string),
            'sz0': tf.FixedLenFeature([], tf.int64),
            'sz1': tf.FixedLenFeature([], tf.int64),
            'depth': tf.FixedLenFeature([], tf.int64),
            'label': tf.FixedLenFeature([], tf.int64)
        }
    )

    tf_clip = tf.decode_raw(features['clip_raw'], tf.float32)#获得图像原始的数据

    tf_sz0 = features['sz0']
    tf_sz1 = features['sz1']
    tf_depth = features['depth']
    tf_label = features['label']

    # PS:恢复原始图像数据,reshape的大小必须与保存之前的图像shape一致,否则出错
    tf_clip=tf.reshape(tf_clip, [sz0, sz1, 3]) # 设置图像的维度


    # 这里仅仅返回图像和标签

    # return tf_image, tf_height,tf_width,tf_depth,tf_label

    return tf_clip,tf_sz0, tf_sz1, tf_label
```

```python
shuffle=False
log=5

# create records file
info_file = 'adobe-nfs-split.txt'
record_output = 'F:\winston\data\\train-70-test-30-clength-50.tfrecords'

create_records(info_file, record_output, shuffle,log)

train_nums=get_example_nums(record_output)

print("save train example nums={}".format(train_nums))
```

```python
def disp_records(record_file, show_nums=1):
    '''
    解析record文件，并显示show_nums张图片，主要用于验证生成record文件是否成功
    :param tfrecord_file: record文件路径
    :return:
    '''
    # 读取record函数
    tf_image, tf_label = read_records(record_file,resize_height,resize_width,type='normalization')

    # 显示前4个图片

    init_op = tf.initialize_all_variables()

    with tf.Session() as sess:

        sess.run(init_op)

        coord = tf.train.Coordinator()

        threads = tf.train.start_queue_runners(sess=sess, coord=coord)

        for i in range(show_nums):

            image,label = sess.run([tf_image,tf_label])  # 在会话中取出image和label

            # image = tf_image.eval()

            # 直接从record解析的image是一个向量,需要reshape显示

            # image = image.reshape([height,width,depth])

            print('shape:{},tpye:{},labels:{}'.format(image.shape,image.dtype,label))

            # pilimg = Image.fromarray(np.asarray(image_eval_reshape))

            # pilimg.show()

            show_image("image:%d"%(label),image)

        coord.request_stop()

        coord.join(threads)
```
