```python
import imageio
import os
import numpy as np
from skimage.io import imsave
```

```python
videoPath = '/home/winston/data/Adobe-240-fps/original_high_fps_videos'
videoList = os.listdir(videoPath)
print(videoList)
```

```python
vidNum = 0
validFormat = ['m4v', 'mp4', 'mov', 'MOV', 'MP4']
savePath = '/home/winston/data/Adobe-240-fps/images'
for vidName in videoList:
    if vidName[-3:] in validFormat:
        print("Processing file %s" % vidName)
        if not os.path.isdir(os.path.join(savePath, vidName[:-4])):
            os.mkdir(os.path.join(savePath, vidName[:-4]))
            vid_reader = imageio.get_reader(os.path.join(videoPath, vidName), 'ffmpeg')
            for i, img in enumerate(vid_reader):
                if i > 100:
                    saveName = os.path.join(savePath, vidName[:-4]) + '/%04d.jpg' % (i-101)
                    imsave(saveName, img)
                    print("--Saved %s" % saveName)
            vidNum += 1
    else:
        print("Not a video file %s" % vidName)
    
print("Processed %d videos" % vidNum)
```
