This script makes a statistical estimation on davis dataset. The goal is to find some empirical number
based on binning scheme 1, e.g. how many events are in one event frame. This distribution is a guidance to synthetic data.

```python
import numpy as np
import utils.events_processing as ep
import utils.visualization as vs
import os
import matplotlib.pyplot as plt
```

```python
def binningStats(eventTxtFile):
    currentFrame = np.zeros((180, 240))
    startTime = 0.0
    durationArray = []
    eventsNumArray = []
    speedArray = []
    for line in open(eventTxtFile, 'r'):
        txyp = line[:-1].split(' ')
        timestamp = float(txyp[0])
        x = int(txyp[1])
        y = int(txyp[2])
        
        if currentFrame[y, x] == 1:
            durationArray.append(timestamp - startTime)
            startTime = timestamp
            eventsNumArray.append(np.sum(currentFrame))
            speedArray.append(eventsNumArray[-1]/durationArray[-1])
            currentFrame = np.zeros((180, 240))
        else:
            currentFrame[y, x] += 1
    print("Finished reading data!")
    return durationArray, eventsNumArray, speedArray
```

```python
dataFolder = r"F:\Winston\data\davis240\urban"
evFilename = os.path.join(dataFolder, 'events.txt')
duration, eventsNum, speed = binningStats(evFilename)

array_x = range(len(duration))

meanPerc = np.mean(eventsNum) /240/180*100
minPerc = np.min(eventsNum) /240/180*100
maxPerc = np.max(eventsNum) /240/180*100
print("mean percentage", meanPerc)
print("min percentage", minPerc)
print("max percentage", maxPerc)
#percHist, binEdges = np.histogram(np.divide(eventsNum,240*180), bins = np.linspace(0,1, 10))
plt.figure(figsize = (10,5))
plt.subplot(221)
plt.plot(array_x, duration)
plt.subplot(222)
plt.plot(array_x, eventsNum)
plt.subplot(223)
plt.plot(array_x, speed)
plt.subplot(224)
plt.hist(np.divide(eventsNum,240*180/100),bins=10)
```
